```python
class batch_GD:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        self.loss_history = []
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # Initialize parameters to zero
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Gradient descent loop
        for epoch in range(self.n_iters):
            # Forward pass
            y_pred = np.dot(X, self.weights) + self.bias
            residual = y_pred - y
            
            # Calculate MSE loss
            mse_loss = np.mean(residual**2)
            
            n = len(y)
            
            # Compute gradients (positive form)
            grad_b = (2/n) * np.sum(residual)
            grad_w = (2/n) * np.dot(X.T, residual)
            
            # Update parameters (subtract gradient)
            self.weights = self.weights - self.learning_rate * grad_w
            self.bias = self.bias - self.learning_rate * grad_b
            
            # Track loss history
            self.loss_history.append(mse_loss)
            
            # Print progress every N epochs
            if (epoch + 1) % 100 == 0:
                print(f"Epoch {epoch+1}, Loss: {mse_loss:.6f}")
        
        print(f"final weights: {self.weights}, bias: {self.bias}")
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
bgd = batch_GD()
bgd.fit(X_train,y_train)
```
