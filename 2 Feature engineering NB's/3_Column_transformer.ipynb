```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['First','Second','Third']])
```

```python
df['class'] = oe.fit_transform(df[['class']])
```

```python
df['age'] = si.fit_transform(df[['age']])
df.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
# ct = ColumnTransformer()
```

```python
df2 = sns.load_dataset('titanic')
```

```python
df2.head()
```

```python
transformer = ColumnTransformer(transformers=[
    ('t1',SimpleImputer(),['age']),
    ('t2',OrdinalEncoder(categories=[['First','Second','Third']]),['class']),
    ('t3',OneHotEncoder(drop='first'),['sex'])
],remainder='drop')
```

```python
df2_data = transformer.fit_transform(df2)
df2_data
```

```python
pd.DataFrame(df2_data,columns=[['age','class','sex']])
```
