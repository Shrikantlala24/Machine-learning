# Covert to datetime

```python
orders['date'] = pd.to_datetime(orders['date'])
orders.info()
```

---
# extracting from datetime
1. year
2. month
3. day
4. month_name() 
5. dayofweek (day of the week)
6. day_name()
7. isocalendar().week (week of the year)
---

```python
orders['year'] = orders['date'].dt.year
orders.head()
```

```python
orders['month'] = orders['date'].dt.month
orders.head()
```

```python
orders['day'] = orders.date.dt.day
orders.head()
```

```python
# orders.drop(columns=['actual_date'],inplace=True)
orders.head()
```

```python
orders['month_name'] = orders.date.dt.month_name()
orders.head()
```

```python
orders['day_of_week'] = orders.date.dt.dayofweek
orders.head()
```

```python
orders['day_name'] = orders.date.dt.day_name()
orders.head()
```

```python
orders['week_of_year'] = orders['date'].dt.isocalendar().week
orders.head()
```
