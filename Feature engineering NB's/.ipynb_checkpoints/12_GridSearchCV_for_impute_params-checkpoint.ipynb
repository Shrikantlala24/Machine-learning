```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.linear_model import LogisticRegression
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.drop(columns=['class','who','adult_male','deck','embark_town','alive','alone'],inplace=True)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['survived']), df['survived'], test_size=0.2,random_state=42)
print(y_test.head())
X_train.head()
```

```python
num = ['age','fare']
cat = ['sex','embarked']
```

# the Pipelines to deal with numerical and categorical features
setting the transformers for pre-processing

```python
# --------------------------------------------------------------------------

num_trf = Pipeline(
    steps=[
        ('impute', SimpleImputer(strategy='median')),
        ('scale', StandardScaler())
    ]
)

# --------------------------------------------------------------------------

cat_trf = Pipeline(
    steps = [
        ('impute', SimpleImputer(strategy='most_frequent')),
        ('ohe',    OneHotEncoder(handle_unknown='ignore'))
    ]
)
```

# the pipeline used in column transformer
> i don't know why this order

```python
pre_processor = ColumnTransformer(
    transformers = [
        ('num',num_trf,num),
        ('cat',cat_trf,cat)
    ]
)
```

# final column transformer - to - pipeline 
> for final model

```python
clf = Pipeline(
    steps = [
        ('pre-processor', pre_processor),
        ('Logistic Regression', LogisticRegression())
    ]
)
```

```python
from sklearn import set_config

set_config(display='diagram')
clf
```
