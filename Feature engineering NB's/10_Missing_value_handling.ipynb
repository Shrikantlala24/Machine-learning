# Numerical Inputation
- Mean/median imputation will be don in age column
- deck has 77% missing values

```python
sns.distplot(df['age'])
```

```python
mean_age = df['age'].mean()
median_age = df['age'].median()
```

```python
df['age_mean'] = df['age'].fillna(mean_age)
df['age_median'] = df['age'].fillna(median_age)
```

```python
sns.distplot(df['age_mean'], color='green')
sns.distplot(df['age_median'],color='orange')
sns.distplot(df['age'])
```
