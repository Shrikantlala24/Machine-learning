```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
# %pip install kagglehub
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rakeshrau/social-network-ads")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/Social_Network_Ads.csv")
df.head()
```

```python
df.drop(['User ID', 'Gender'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
X_test.shape
```

```python
X_train.head()
```

```python
sc.fit(X_train)

print(sc.mean_)
print(sc.scale_)
print(sc.var_)
```

```python
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# X_train_scaled.describe()

# ! the error is because the sc.transform converts the DF to an ndarray
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_train_scaled.describe().round()
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_test_scaled.describe().round()
```

### effect of standardization

```python
import matplotlib.pyplot as plt
```

```python
fig ,(ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'], color='blue')
ax1.set_title('Original Data')


ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], color='red')
ax2.set_title('Standardized Data')
```

```python
import seaborn as sns
```

```python
fig, (a1, a2) = plt.subplots(1, 2, figsize=(12, 5))

sns.kdeplot(X_train['Age'], ax=a1, color='blue', label='Age')
sns.kdeplot(X_train_scaled['Age'], ax=a2, color='blue', label='Age')
a1.set_title('Original Data')

a2.set_title('Standardized Data')

sns.kdeplot(X_train['EstimatedSalary'], ax=a1, color='red', label='EstimatedSalary')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=a2, color='red', label='EstimatedSalary')
a1.legend()
a2.legend()
```
