### Probablity distributions

here we'll discuss all the codes relevant to probability and stats

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\students_normal_distribution.csv")
df
```

```python
plt.hist(df['Marks'], bins=10, edgecolor='black')
plt.title('Distribution of Student Scores')

plt.xlabel('Marks')
plt.ylabel('Number of Students')
```

```python
marks = df['Marks']
mean_marks = np.mean(marks)
std_dev_marks = np.std(marks)
```

```python
# !pip install scipy
# already installed in the environment
```

```python

# we are using scipy to create a normal distribution

from scipy.stats import norm
dist = norm(mean_marks, std_dev_marks)

x = np.linspace(min(marks), max(marks), 100)
paltudf = dist.pdf(x)
```

```python
# plt.hist(marks, bins=10, density=True)
plt.plot(x,paltudf, color='red')
```

```python
plt.hist(marks, bins=10, density=True)
plt.plot(x,paltudf, color='red')
```
