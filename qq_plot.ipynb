```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.kdeplot(data = df['sepal_length'])
```

```python
# Let's create the Q-Q plot

# [X,Y]
x_quant = []

x = sorted(df['sepal_length'].tolist())
for i in range(1,101):
    x_quant.append(np.percentile(x, i))
```

```python
y_quant = []
y = np.random.normal(loc=np.mean(x), scale=np.std(x), size=1000)
y = sorted(y.tolist())
for i in range(1,101):
    y_quant.append(np.percentile(y, i))


```
