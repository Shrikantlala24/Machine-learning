```python
# Generate a synthetic dataset with 1000 samples and save as CSV

import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

data = {
    "Student_ID": np.arange(1, n+1),
    "Study_Hours": np.random.randint(1, 11, n),
    "Sleep_Hours": np.random.randint(4, 9, n),
    "Attendance": np.random.randint(50, 101, n),
    "Previous_Grade": np.random.randint(40, 91, n),
    "Assignments_Completed": np.random.randint(1, 11, n),
    "Screen_Time": np.random.randint(1, 9, n)
}

df = pd.DataFrame(data)

# Create Final Score with some realistic relationship
df["Final_Score"] = (
    0.3 * df["Study_Hours"] * 10 +
    0.2 * df["Attendance"] +
    0.2 * df["Previous_Grade"] +
    0.2 * df["Assignments_Completed"] * 10 -
    0.2 * df["Screen_Time"] * 5 +
    np.random.normal(0, 5, n)
)

df["Final_Score"] = df["Final_Score"].clip(0, 100).round(0)

# file_path = "/content/sample_data/student_performance_1000_samples.csv"
# df.to_csv(file_path, index=False)

# file_path
```

```python
df.head()
```

```python
plt.plot(df["Study_Hours"], df["Final_Score"], "_")

# Que :
# 1 .what's the relation between the student marks and thier study time
# thier could be may questions

# answers
# 1. as the hours (^) , the score(^)
# 2. top performer is not spending 10 hrs (highest pt at 8 hrs)

plt.xlabel('Study time (in HRS)')
plt.ylabel('Exam score (out of 100)')
plt.title('Study time vs Exam scores')
```
